````python
import pandas as pd
import numpy as np
import scipy.stats as s
````

````python
from sklearn.decomposition import PCA
````

````python
data = pd.read_csv("/kaggle/input/laptop-price/laptop_price.csv",encoding="latin-1")
````

````python
data.drop(labels=data.columns[0],axis=1,inplace=True)
````

````python
company_ohe = pd.crosstab(data.index,data["Company"])
data.drop(labels=["Company"],axis=1,inplace=True)
data = pd.concat([company_ohe,data],axis=1)
````

````python
price_groupby_type = data.groupby(by="TypeName")
price_dict = price_groupby_type["Price_euros"].agg("mean")
price_dict.sort_values()
ordinal_mapping_type = dict(zip(price_dict.index,range(len(price_dict))))
data["TypeName"] = data["TypeName"].map(ordinal_mapping_type)
````

````python
price_groupby_res = data.groupby(by="ScreenResolution")
price_dict = price_groupby_res["Price_euros"].agg("mean")
price_dict.sort_values()
ordinal_mapping_res = dict(zip(price_dict.index,range(len(price_dict))))
data["ScreenResolution"] = data["ScreenResolution"].map(ordinal_mapping_res)
````

````python
price_groupby_cpu = data.groupby(by="Cpu")
price_dict = price_groupby_cpu["Price_euros"].agg("mean")
price_dict.sort_values()
ordinal_mapping_cpu = dict(zip(price_dict.index,range(len(price_dict))))
data["Cpu"] = data["Cpu"].map(ordinal_mapping_cpu)
````

````python
data["Ram"] = data["Ram"].apply(lambda x: int(x.split("GB")[0]))
````

````python
data["Weight"] = data["Weight"].apply(lambda x: float(x.split("kg")[0]))
````

````python
price_groupby_mem = data.groupby(by="Memory")
price_dict = price_groupby_mem["Price_euros"].agg("mean")
price_dict.sort_values()
ordinal_mapping_mem = dict(zip(price_dict.index,range(len(price_dict))))
data["Memory"] = data["Memory"].map(ordinal_mapping_mem)
````

````python
gpu_ohe = pd.crosstab(data.index,data["Gpu"])
data.drop(labels=["Gpu"],axis=1,inplace=True)
data = pd.concat([gpu_ohe,data],axis=1)
````

````python
os_ohe = pd.crosstab(data.index,data["OpSys"])
data.drop(labels=["OpSys"],axis=1,inplace=True)
data = pd.concat([os_ohe,data],axis=1)
````

````python
product_ohe = pd.crosstab(data.index,data["Product"])
data.drop(labels=["Product"],axis=1,inplace=True)
data = pd.concat([product_ohe,data],axis=1)
````

### To get more input features which have high corrleation with the target feature of ```Price_euros```, you can actually perform Pearson Correlation Test on each of the input features with the Target Feature like this:

````python
refined_columns = list()

for input_feature_column in data.columns:

    two_tailed_pearsonr_result = s.pearsonr(x=data[input_feature_column],y=data[data.columns[-1]])
    if two_tailed_pearsonr_result.pvalue < 0.01:

        refined_columns.append(input_feature_column)
````

````python
len(refined_columns)
````

### As you can see that while setting our own course value of Pearson Corrleation as $\pm 0.3, \pm 0.4 \text{or} \pm 0.5$, it's better to perform Pearson Correlation test as it is is giving us far more greater number of Input Features (79) in this case.

````python
refined_data = data[refined_columns]
````

````python
sigma = np.cov(refined_data.iloc[:,0:-1],rowvar=False)
````

````python
X = refined_data.iloc[:,0:-1]
pca = PCA(n_components=12)

X_transformed = pca.fit_transform(X)
````

````python
transformed_X = pd.DataFrame(X_transformed,columns=["feat_"+str(i) for i in range(1,13)])
````
